#### 트리의 구조와 관련 용어
트리 구성 :
- 노드 node & 가지 edge
- 각 노드는 가지를 통해 다른 노드와 연결됨


##### 루트 root
- 가장 위쪽에 있는 노드
- 트리 하나에 루트는 1개만 존재

##### 리프 leaf
- 가장 아래 쪽에 있는 노드
- also known as 단말 노드 terminal node, 외부 노드 external node
- 물리적으로 가장 밑에 위치한다는 의미가 아닌 가지가 더 이상 뻗어나갈 수 없는 마지막에 노드가 있다는 뜻

##### 비단말 노드 (non-terminal node / internal node)
- 리프를 제외한 노드 (루트 포함)

##### 자식 child
- 어떤 노드와 가지가 연결되었을 때 아래쪽 노드를 자식 child 라고 함
- one to many

##### 부모 parent
- 어떤 노드와 가지가 연결되었을 때 위쪽에 있는 노드를 부모 parent 라고 함
- 노드의 부모는 하나

##### 형제 sibling
- 부모가 같은 노드를 형제 sibling 라고 함

##### 조상 ancestor
- 어떤 노드에서 위쪽으로 가지를 따라가면 만나는 모든 노드

##### 자손 descendant
- 어떤 노드에서 아래쪽으로 가지를 따라가면 만나는 모든 노드

##### 레벨 level
- 루트에서 얼마나 멀리 떨어져 있는지를 나타내는 것이 레벨 level
- 루트 레벨 : 0
- 가지가 하나씩 아래로 뻗어 내려갈 때마다 레벨이 1씩 증가

##### 차수 degree
- 각 노드가 갖는 자식의 수 / how many children
- 모든 노드의 차수가 n 이하인 트리를 n진 트리라고 함
- 자식이 3이 최대일 경우 > 3진 트리

##### 높이 height
- 루트에서 가장 멀리 있는 리프까지의 거리
- 리프 레벨의 최댓값

##### 서브트리 subtree
- 루트가 있는 트리에서, 그 자손으로 구성된 트리 (그 자손이 해당 서브트리에서 루트가 됨)

##### 빈 트리 None tree / Null tree
- 노드와 가지가 전혀 없는 트리

#### 순서 트리 ordered tree 의 검색
공부 일자 : 2025-12-03

크게 2가지 방법
- 너비 우선 vs 깊이 우선


##### 너비 우선 검색 breadth-first search
- 폭 우선, 가로 검색, 수평 검색
- 낮은 레벨부터 (from the top/root) 왼쪽에서 오른쪽으로 검색
- 한 레벨에서 검색을 마치면 다음 레벨로 내려가는 방법

##### 깊이 우선 검색 depth-first search
- 세로 검색, 수직 검색
- 리프에 도달할 때까지 아래쪽으로 내려가면서 검색하는 것을 우선으로 하는 방법
- 리프에 도달 시, 일단 부모 노드로 돌아가서 가서 다음 자식 노드로 내려감


  ###### 전위 순회 preorder (루트로 시작)
  - 루트 > 왼쪽 자식 > 오른쪽 자식

  ###### 중위 순회 inorder (루트가 중앙)
  - 왼쪽 리프에서 시작 > 루트 > 오른쪽 (중에서 왼쪽) 리프에서 시작

  ###### 후위 순회 postorder (루트가 끝)
  - 왼쪽 리프 > 레벨 하나만 올라갔다가 다시 원래 리프 레벨단에 없는지 검색 > 그런담에 왼쪽 첫번째 자식 까지 올라갔다가 > 다시 오른쪽 리프 (중에서 가장 왼쪽에 있는 리프부터)

#### 이진 트리 알아보기 binary tree
- 노드가 왼쪽 자식 left child 과 오른쪽 자식 right child 만 갖는 트리
- 두 자식 가운데 하나 또는 둘 다 존재하지 않는 노드가 있어도 괜찮음

**특성**
- 왼쪽 자식과 오른쪽 자식을 구분한다는 점
- 왼쪽 자식을 루트로 하는 서브트리 : left subtree
- 오른쪽 자식을 루트로 하는 서브트리 : right subtree


#### 완전 이진 트리 알아보기 complete binary tree
- 루트부터 아래쪽 레벨로 노드가 가득 차 있고, 같은 레벨 안에서 왼쪽부터 오른쪽으로 노드가 채워져 있는 이진 트리
- 마지막 레벨을 제외 하고 모든 레벨에 노드가 가득 차 있음
- 마지막 레벨에 한해서 왼쪽부터 오른쪽으로 노드를 채우되 반드시 끝까지 채우지 않아도 됨
