### 08-2 포인터를 이용한 연결 리스트
공부일자: 2025-11-25



#### **class Node**
- __init__() 함수가 있음

#### **field**
class Node 는 2개의 필드로 구성
1. data - 데이터에 대한 참조 : 임의의 형
2. next - 뒷쪽 노드에 대한 참조 : Node 형

```python
# 포인터로 연결 리스트 구현하기

from __future__ import annotations
from typing import Any, Type

class Node :
  """ 연결 리스트용 노드 클래스"""

  def __init__(self, data : Any = None, next : Node = None):
    """초기화"""
    self.data = data #데이터
    self.next = next #뒤쪽 포인터
```

보충수업 : 파이썬의 리스트는 자료 구조가 아닙니다

- 연결 리스트의 장점 : 임의의 위치에 원소를 삽입하거나 삭제할 때 빠르게 수행 할 수 있음
- 단점 : 메모리 (기억 영역) 과 속도 면에서는 배열 보다 효율이 떨어짐

파이썬의 리스트
- 연결 리스트의 자료 구조가 아니라 모든 원소를 연속으로 메모리에 배치하는 '배열'로 내부에서 구현함

노드를 스캔할 때 다음 조건 가운데 하나만 성립해도 검색이 종료됨
- 종료조건 1 : 검색 조건을 만족하는 노드를 발견하지 못하고 꼬리 노드까지 왔을 경우
- 종료조건 2 : 검색 조건을 만족하는 노드를 발견한 경우

```python
def search(self, data : Any) -> int:
  """ data와 값이 같은 노드를 검색"""
  cnt = 0
  ptr = self.head
  while ptr is not None:
    if ptr.data == data :
      self.current = ptr
      return cnt
    cnt += 1
    ptr = ptr.next
  return -1

def __contains__(self, data : Any) -> bool :
  """ 연결 리스트에 data 가 포함되어 있는 지 확인 """
  return self.search(data) >= 0
```

데이터가 포함되어 있는 지 판단하는 `__contains__()` 함수
- 리스트에 data 와 값이 같은 노드가 포함되어 있는 지 판단 하는 함수
- 포함되어 있으면 True / 없으면 False
