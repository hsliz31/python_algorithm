### 08-2 포인터를 이용한 연결 리스트
공부일자: 2025-11-25, 26

- 노드를 삽입-삭제할 때 `데이터를 이동하지 않고 처리` 하는 특징
- 노드를 삽입-삭제랄 때마다 노드용 인스턴스를 생성하고 소멸
  - 메모리 확보/해제 하는데 쓰는 비용이 큼




#### **class Node**
- __init__() 함수가 있음

#### **field**
class Node 는 2개의 필드로 구성
1. data - 데이터에 대한 참조 : 임의의 형
2. next - 뒷쪽 노드에 대한 참조 : Node 형

```python
# 포인터로 연결 리스트 구현하기

from __future__ import annotations
from typing import Any, Type

class Node :
  """ 연결 리스트용 노드 클래스"""

  def __init__(self, data : Any = None, next : Node = None):
    """초기화"""
    self.data = data #데이터
    self.next = next #뒤쪽 포인터
```

보충수업 : 파이썬의 리스트는 자료 구조가 아닙니다

- 연결 리스트의 장점 : 임의의 위치에 원소를 삽입하거나 삭제할 때 빠르게 수행 할 수 있음
- 단점 : 메모리 (기억 영역) 과 속도 면에서는 배열 보다 효율이 떨어짐

파이썬의 리스트
- 연결 리스트의 자료 구조가 아니라 모든 원소를 연속으로 메모리에 배치하는 '배열'로 내부에서 구현함

노드를 스캔할 때 다음 조건 가운데 하나만 성립해도 검색이 종료됨
- 종료조건 1 : 검색 조건을 만족하는 노드를 발견하지 못하고 꼬리 노드까지 왔을 경우
- 종료조건 2 : 검색 조건을 만족하는 노드를 발견한 경우

```python
def search(self, data : Any) -> int:
  """ data와 값이 같은 노드를 검색"""
  cnt = 0
  ptr = self.head
  while ptr is not None:
    if ptr.data == data :
      self.current = ptr
      return cnt
    cnt += 1
    ptr = ptr.next
  return -1

def __contains__(self, data : Any) -> bool :
  """ 연결 리스트에 data 가 포함되어 있는 지 확인 """
  return self.search(data) >= 0
```

데이터가 포함되어 있는 지 판단하는 `__contains__()` 함수
- 리스트에 data 와 값이 같은 노드가 포함되어 있는 지 판단 하는 함수
- 포함되어 있으면 True / 없으면 False

1. 머리에 노드를 삽입하는 add_first() 함수
2. 꼬리에 노드를 삽입하는 add-last() 함수
3. 머리 노드를 삭제하는 remove_first() 함수
4. 꼬리 노드를 삭제하는 remove_last() 함수
5. 임의의 노드를 삭제하는 remove() 함수
6. 주목 노드를 삭제하는 remove_current_node() 함수
7. 모든 노드를 삭제하는 clear() 함수
8. 주목 노드를 한 칸 뒤로 이동시키는 next() 함수
9. 주목 노드를 출력하는 print_current_node() 함수
10. 모든 노드를 출력하는 print() 함수

### 08-3 커서를 이용한 연결 리스트

커서 (cursor) : 뒤쪽 포인터 / 뒤쪽 노드가 저장되는 원소의 `인덱스` / 인덱스로 나타낸 포인터
- 꼬리노드의 뒤쪽 커서는 -1

`ArrayLinkedList`

#### **배열 안에 비어 있는 원소 처리하기**
- 배열에서 물리적인 위치 관계와 연결 리스트의 논리적인 순서 관계가 일치하는 것은 아님
- 즉, 리스트에서 n 번째 위치하는 노드가 배열 인덱스 n의 원소에 저장되는 것은 아님

|:---|:---|:--- |:--- |
| 0 | C | 2 | |
| 1 | A | 3 | |
| 2 | D | -1 | 제일 끝|
| 3 | B | 0 | |
| 4 | E | 1 | 제일 앞 |
| 5 | - | - |

| 인덱스 | 노드 | 뒤쪽 노드 인덱스 | 비고 |
|:---|:---|:--- |:--- |
| 0 | C | 2 | |
| 1 | A | 3 | |
| 2 | D | -1 | 제일 끝|

- 삭제를 여러 번 반복하면 배열 안에 빈 레코드가 많이 생김
- 삭제되는 레코드가 딱 하나라면 그 인덱스를 어떤 변수에 넣어 놓고 관리함으로써 쉽게 재사용 할 수 있음

#### **프리 리스트 (free list)**

- 연결 리스트
- 삭제된 레코드 그룹을 관리할 때 사용하는 자료구조

**추가 되는 필드**
- 노트 클래스 `Node`에 추가된 필드

- 연결 리스트 클래스 `ArrayLinkedList` 에 추가된 필드
  - `deleted` : 프리 리스트의 머리 노드를 참조하는 커서
  - `max` : 배열에서 맨 끝 쪽에 저장되는 노드의 레코드 번호
